```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Placeholder for df. You should replace this with your actual data loading.
# For example: df = pd.read_csv('your_data.csv')
df = pd.DataFrame({
    'Rainfall_mm': [100, 150, 200, 250, 300],
    'Yield_kg': [50, 75, 100, 125, 150]
})

X = df[["Rainfall_mm"]]
y = df["Yield_kg"]
# Very simple model
model_simple = LinearRegression()
model_simple.fit(X, y)
model_simple.predict([[205]])
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
# Very complex model
model_complex = make_pipeline(
 PolynomialFeatures(degree=5),
 LinearRegression()
)
model_complex.fit(X, y)
model_complex.predict([[205]])
```
