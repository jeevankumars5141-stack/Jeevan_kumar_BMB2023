```python

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# 1. Generate Fake Data
data = pd.DataFrame(np.random.rand(100, 3), columns=['Nitrogen','Rainfall','Temperature'])
data['Yield'] = data.sum(axis=1) * 10 + np.random.normal(0, 5, 100)

# 2. Train
X = data.drop('Yield', axis=1)
y = data['Yield']
model = RandomForestRegressor().fit(X, y)
print(f"Model trained to predict Yield. R2 Score: {model.score(X, y):.2f}")
```
