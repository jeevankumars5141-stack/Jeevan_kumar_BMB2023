```python

# Simple AI Decision System for Irrigation
# This demonstrates how AI "thinks" about farming decisions

def irrigation_advisor(soil_moisture, temperature, forecast_rain):
    """
    A rule-based AI system for selection of crop decisions.
    This mimics how AI processes multiple factors.

    Parameters:
    - soil_moisture: percentage (0-100)
    - temperature: degrees Celsius
    - forecast_rain: True/False
    - soil_condition: clay/sandy/loamy/silty
    - crop_variety: newly_released
    - market demand: consumer_income/tastes/preferences
    """

    # AI "reasoning" - considering multiple factors
    if soil_moisture < 30:
        if forecast_rain:
            return "⏳ Wait - Rain expected. Soil is dry but rain will help."
        else:
            return "💧 IRRIGATE NOW - Soil critically dry, no rain expected."

    elif soil_moisture < 60:
        if temperature > 35:
            return "💧 IRRIGATE - High temp causing rapid evaporation."
        else:
            return "✅ OKAY - Moisture adequate for now."

    else:
        return "🛑 NO IRRIGATION NEEDED - Soil has enough moisture."

# Test the AI advisor
print("=== Smart Irrigation Advisor ===")
print(irrigation_advisor(25, 32, False))  # Dry soil, no rain
print(irrigation_advisor(25, 32, True))   # Dry soil, rain coming
print(irrigation_advisor(45, 38, False))  # Medium moisture, hot day
print(irrigation_advisor(75, 28, False))  # Good moisture
```

```python
# Simple AI Decision System for Crop Selection
# This shows how AI helps farmers choose a suitable crop

def crop_advisor(soil_type, rainfall, temperature, season):
    """
    A rule-based AI system for crop selection.

    Parameters:
    - soil_type: 'sandy', 'loamy', 'clayey'
    - rainfall: 'low', 'medium', 'high'
    - temperature: degrees Celsius
    - season: 'kharif', 'rabi', 'summer'
    """

    # AI "reasoning" based on conditions
    if season == "kharif":
        if rainfall == "high":
            return "🌾 Recommended Crop: Rice – Needs more water."
        elif rainfall == "medium":
            return "🌽 Recommended Crop: Maize – Moderate rainfall suitable."
        else:
            return "🌾 Recommended Crop: Millets – Low rainfall tolerant."

    elif season == "rabi":
        if temperature < 25:
            return "🌾 Recommended Crop: Wheat – Cool climate crop."
        else:
            return "🌱 Recommended Crop: Chickpea – Tolerates mild heat."

    elif season == "summer":
        if soil_type == "sandy":
            return "🥜 Recommended Crop: Groundnut – Suitable for sandy soil."
        else:
            return "🌻 Recommended Crop: Sunflower – Heat tolerant."

    else:
        return "❓ Insufficient data to recommend a crop."

# Test the Crop Advisor
print("=== Smart Crop Selection Advisor ===")
print(crop_advisor("clayey", "high", 30, "kharif"))
print(crop_advisor("loamy", "medium", 22, "rabi"))
print(crop_advisor("sandy", "low", 35, "summer"))
```
