```python

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# 1. Generate Fake Data
data = pd.DataFrame(np.random.rand(100, 3), columns=['Humidity','Pressure','WindDir'])
data['Rain_Probability'] = data.sum(axis=1) * 10 + np.random.normal(0, 5, 100)

# 2. Train
X = data.drop('Rain_Probability', axis=1)
y = data['Rain_Probability']
model = RandomForestRegressor().fit(X, y)
print(f"Model trained to predict Rain_Probability. R2 Score: {model.score(X, y):.2f}")
```
